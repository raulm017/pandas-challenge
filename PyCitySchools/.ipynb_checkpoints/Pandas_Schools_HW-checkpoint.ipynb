```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
Total_Schools=school_data_complete["school_name"].nunique()
Total_Schools
```

```python
Total_students=school_data_complete["Student ID"].nunique()
Total_students
```

```python
Total_budget=school_data["budget"].sum()
Total_budget
```

```python
Average_math_score=school_data_complete["math_score"].mean()
Average_math_score
```

```python
Average_reading_score=school_data_complete["reading_score"].mean()
Average_reading_score
```

```python
Passing_math=len(school_data_complete[school_data_complete["math_score"]>=70])/Total_students
Passing_math
```

```python
Passing_reading=len(school_data_complete[school_data_complete["reading_score"]>=70])/Total_students
Passing_reading
```

```python
Passing_math_and_reading=len(school_data_complete[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)])/Total_students
Passing_math_and_reading
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score

  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_data = pd.read_csv(school_data_to_load)
```

```python
school_data["Per Student Budget"]=school_data["budget"]/school_data["size"]
school_data
```

```python
averages=school_data_complete.groupby("school_name")["math_score","reading_score"].mean()
averages=averages.reset_index()

school_data
```

```python
school_data=pd.merge(school_data,averages)
```

```python
school_data_complete["math_ph"]=np.where(school_data_complete["math_score"]>=70,1,0)
school_data_complete
```

```python
math_pass=school_data_complete.groupby("school_name",as_index=False)["math_ph"].sum()
school_data=pd.merge(school_data,math_pass)
school_data["% Passing Math"]=school_data["math_ph"]/school_data["size"]
school_data=school_data.drop("math_ph",1)
school_data
```

```python
school_data_complete["reading_ph"]=np.where(school_data_complete["reading_score"]>=70,1,0)
reading_pass=school_data_complete.groupby("school_name",as_index=False)["reading_ph"].sum()
school_data=pd.merge(school_data,reading_pass)
school_data["% Passing Reading"]=school_data["reading_ph"]/school_data["size"]
school_data=school_data.drop("reading_ph",1)
```

```python
school_data=school_data.drop("School ID",1)
school_data
```

```python
school_data_complete["overall_ph"]=np.where(((school_data_complete["math_ph"]==1) & (school_data_complete["reading_ph"]==1)) ,1,0)
overall_pass=school_data_complete.groupby("school_name",as_index=False)["overall_ph"].sum()
school_data=pd.merge(school_data,overall_pass)
school_data["% Overall Passing"]=school_data["overall_ph"]/school_data["size"]
school_data=school_data.drop("overall_ph",1)
school_data=school_data.rename(columns={"type":"School Type","size":"Total Students","budget":"Total School Budget",
                                        "math_score":"Average Math Score","reading_score":"Average Reading Score"})

```

```python
school_data=school_data.set_index("school_name")
school_data
```
